```python
from pyod.models.knn import KNN
```

```python
from pyod.utils.data import generate_data
```

```python
contamination = 0.1
n_train = 200
n_test = 100
X_train, y_train, X_test, y_test = generate_data(n_train=n_train, n_test=n_test, contamination=contamination)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_train[:,0], X_train[:, 1])
```

```python
plt.scatter(X_test[:,0], X_test[:,1])
```

```python
clf = KNN()
clf.fit(X_train)

y_train_pred = clf.labels_
y_train_scores = clf.decision_scores_

y_test_pred = clf.predict(X_test)
y_test_scores = clf.decision_function(X_test)
```

```python
from pyod.utils.data import evaluate_print
```

```python
print('\nOn Training Data:')
evaluate_print('KNN', y_train, y_train_scores)
print('\nOn Test Data:')
evaluate_print('KNN', y_test, y_test_scores)
```

```python
from pyod.utils.example import visualize
```

```python
visualize('KNN', X_train, y_train, X_test, y_test, y_train_pred, y_test_pred, show_figure=True, save_figure=False)
```
